```python
import pandas as pd



```

```python

dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
print(dataset.columns)
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
input = dataset[['age', 'sex_male', 'bmi','smoker_yes','children']]
```

```python
input
```

```python
output=dataset[['charges']]
```

```python
output
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(input,output,test_size=0.30,random_state=0)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.svm import SVR
regressor = SVR(kernel="poly",C=0.1,degree=3)
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test,y_pred)
```

```python
r_score
```

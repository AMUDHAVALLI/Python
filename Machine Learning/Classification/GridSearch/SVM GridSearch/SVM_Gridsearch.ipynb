```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
inputdata = dataset[['age', 'sex_male', 'bmi','smoker_yes','children']]
```

```python
inputdata
```

```python
output=dataset[['charges']]
```

```python
output
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(inputdata,output,test_size=0.30,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train= sc.fit_transform(x_train)
x_text=sc.transform(x_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={'kernel':['rbf', 'poly', 'sigmoid', 'linear'],
            'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(x_train,y_train)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter {}:".format(grid.best_params_))
```

```python
table= pd.DataFrame.from_dict(re)
```
